### Data preprocessing

There are quite a lot of missing values in the data, there are columns with an incorrect data type. There are gaps in all columns except total_images, last_price, total_area, first_day_exposition, rooms, floor, studio, open_plan.     

```python
data = data.rename(columns={'balcony' : 'balcony', 'cityCenters_nearest' : 'city_centers_nearest'})
```

```python
print(data['ceiling_height'].unique())
```

```python
data.loc[(data['ceiling_height'] > 20) & (data['ceiling_height'] < 50), 'ceiling_height'] = \
data['ceiling_height'] / 10
```

```python
#removing apartments with ceilings below 2.5 m, because according to the SNIP, the minimum height is 2.5 meters,
#and more than 10 meters, since such objects are more likely to be anomalous.
data['ceiling_height'] = data.loc[(data['ceiling_height'] > 2.5) & (data['ceiling_height'] < 10), 'ceiling_height']
print(data['ceiling_height'].unique())
```

```python
#filling in the gaps in the ceiling_height column with the median value, because the average will be incorrect
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
print(data['ceiling_height'].unique())
```

```python
#filled in the gaps in the balcony column, replaced them with 0,
#because we assume that if there is no information about the balcony, it means that there is no balcony in the apartment.
#changing the column type to integer
data['balcony'] = data['balcony'].fillna('0')
data['balcony'] = data['balcony'].astype('int')
print(data['balcony'].value_counts())
```

```python
data.loc[data['balcony'] > data['rooms'], 'balcony'] = data['rooms'].mean()
print(data['balcony'].unique())
```

```python
#removing outliers in the floors_total column
data['floors_total'] = data.loc[data['floors_total'] <29, 'floors_total']
print(data['floors_total'].median())
data.isna().sum()
data['floors_total'] = data['floors_total'].fillna(data['floors_total'].median())
data['floors_total'] = data['floors_total'].astype('int')
print(data['floors_total'].value_counts())
```

Removing apartments with a living area and a kitchen area of less than 8 sq. m., such a thing contradicts the norms of the SNIP, we also get rid of objects where the living area is more than 250 sq.m. and the kitchen area is more than 60 sq. m., because when studying histograms, it can be noticed that objects with a similar quadrature are rather anomalies and can distort the study

```python
data['total_area'] = data.loc[data['total_area'] <= 400, 'total_area']
data['living_area'] = data.loc[(data['living_area'] > 8) & (data['living_area'] <= 250), 'living_area']
data['kitchen_area'] = data.loc[(data['kitchen_area'] > 8) & (data['kitchen_area'] <= 60), 'kitchen_area']
```

```python
print(data['total_area'].sort_values())
```

```python
#Removing apartments with more than 9 rooms, because if you look at the data, 
#it is rather abnormal data that can spoil the study, and we change the omissions and 0 to 1, 
#because in any case there is at least one room in the apartment


#data['rooms'] = data.loc[data['rooms'] <= 5, 'rooms']
#data.loc[data['rooms'] == 0, 'rooms'] = 1
#data['rooms'] = data['rooms'].fillna(1)
#print(data['rooms'].unique())
```

```python
data['rooms'].hist()
```

Removing the outlier - an apartment with a price of 12 thousand

```python
data['last_price'] = data.loc[data['last_price'] > 13000, 'last_price']
```

```python
#changing the data type of the is_apartment column, the gaps disappear immediately
data['is_apartment'] = data['is_apartment'].astype('bool')
print(data['is_apartment'])
print(data['is_apartment'].unique())
```

```python
#replacing all duplicates
data['locality_name'] = data['locality_name'].str.lower().str.replace('ё', 'е')
data['locality_name'] = data['locality_name'].str.replace('поселок городского типа', 'городской поселок')
data['locality_name'] = data['locality_name'].str.replace('городской поселок', 'поселок')
data['locality_name'] = data['locality_name'].str.replace('поселок мурино', 'мурино')
data['locality_name'] = data['locality_name'].str.replace('село рахья', 'рахья')
data['locality_name'] = data['locality_name'].str.replace('поселок гарболово', 'деревня гарболово')
data['locality_name'] = data['locality_name'].str.replace('поселок зимитицы', 'деревня зимитицы')
data['locality_name'] = data['locality_name'].str.replace('поселок калитино', 'деревня калитино')
data['locality_name'] = data['locality_name'].str.replace('деревня кузьмолово', 'поселок кузьмоловский')
data['locality_name'] = data['locality_name'].str.replace('поселок павлово', 'деревня павлово')
data['locality_name'] = data['locality_name'].str.replace('поселок рабитицы', 'деревня рабитицы')
data['locality_name'] = data['locality_name'].str.replace('поселок при железнодорожной станции вещево', 'поселок станции вещево')
data['locality_name'] = data['locality_name'].str.replace('поселок при железнодорожной станции приветнинское', 'поселок станции приветнинское')
data['locality_name'] = data['locality_name'].str.replace('поселок федоровское', 'деревня федоровское')
print(data['locality_name'].unique())
```

The data was analyzed and anomalies were found in them. So, real estate with an exceptionally low value (about 12 thousand) was excluded. Apartments with ceilings with a height of less than 2.5 meters and more than 10 meters were excluded, the gaps were filled with a median value. Objects with more than 29 floors were also excluded, firstly, St. Petersburg is not a city with a huge number of multi-storey buildings and some of the objects in the dataset were simply not plausible, while others excluded were simply anomalous and could harm the results of the study. The names of localities were analyzed and duplicates were removed. Objects whose area contradicted the norms of the SNIP were also removed.

```python
#adding a column with the price of one square meter 
data['metr_price'] = data['last_price']/data['total_area']
print(data['metr_price'])
```

```python
data['weekday'] = data['first_day_exposition'].dt.weekday
data['month'] = data['first_day_exposition'].dt.month
data['year'] = data['first_day_exposition'].dt.year
print(data.head())
```

```python
def floor_type(row):
    try:
        if row['floor'] == 1:
            return 'первый'
        elif row['floor'] == row['floors_total']:
            return 'последний'
        elif row['floor'] == row['floor'] & row['floor'] != 1 & row['floor'] != row['floors_total']: 
            return 'другой'
    except:
        pass
```

```python
data['floor_type'] = data.apply(floor_type, axis=1)
print(data['floor_type'].unique())
```

```python
data['city_centers_nearest_km'] = data['city_centers_nearest']//1000
print(data['city_centers_nearest_km'])
```

### Conclusion

During the study, it was revealed that most apartments have an area of 25 to 75 sq.m., living area from 8 to 25 sq.m., kitchens - 8-13 sq.m. with ceilings from 2.5 to 2.7 m. Most apartments are located on intermediate between the first and last floors.
Apartments are most often sold in 6 months.
As for the relationship between the indicators and the price of the apartment, further conclusions were obtained:
- the relationship between the price and the total area is clearly present, the correlation coefficient is 0.625
- despite the correlation coefficient, the graph shows that the more rooms in the apartment, the higher the price for the apartment
- when studying the graph on the distribution of apartment prices by year, you can see that prices have fallen significantly since 2014
- the further from the center the apartment is, the cheaper its price
